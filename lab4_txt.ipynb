## 1

```python
import re
f = open("row.txt")
pattern = re.compile(r".*аб*.*")
for i in f.readlines():
    for j in i.split():
        print(j)
        match = pattern.match(j)
        if match: print(match.group())
        else: print("No such word")
```

## 3

```python
import re
f = open("row.txt")
pattern = re.compile(r"\b(?:(?:[а-я]+_)+[а-я]+)\b")
for i in f.readlines():
    print(i)
    if pattern.findall(i) != []: print(pattern.findall(i))
    else: print("No such word")
```

```python
import re
f = open("row.txt")
pattern = re.compile(r"\.+|\W|,+")
for i in f.readlines():
    print(i)
    print(pattern.sub(":",i))
```

## 7

```python
import re
f = open("row.txt")
pattern = r"_([а-я])"
for i in f.readlines():
    for j in i.split():
        print(j)
        result=j[0].upper() + re.sub(pattern,lambda match: match.group(1).upper(),j[1:])
        print(result)
```

## 8

```python
import re
f = open("row.txt")
def split_upp(txt):
    pattern = r"[А-Я]"
    return re.split(pattern, txt)[1:]
for i in f.readlines():
    for j in i.split():
        print(j)
        print(split_upp(j))
```

## 9

```python
import re
f = open("row.txt")
def spaces(txt):
    pattern = r"[А-Я][а-я]*"
    return " ".join(list(re.findall(pattern, txt)))
for i in f.readlines():
    for j in i.split():
        print(j)
        print(spaces(j))
```
